```python
# import packages

import pandas as pd
import warnings
```

```python
# hide warnings 
warnings.filterwarnings('ignore')
```

```python
# read data using pandas
data = pd.read_csv('data/udemy_course_data.csv')
```

```python
# view data
data.head()
```

```python
# check the technical summary
data.info()
```

```python
# check the datatype
data.dtypes
```

```python
# check the shape
data.shape
```

```python
# find the rows which have 'python' keyword in course_title column
data_contain_python_keyword = data[data.course_title.str.contains('python',case=False)]
data_contain_python_keyword.head(5)
```

```python
# find the rows which have 'algo trading' string in course_title column
data_contain_single_string = data[data.course_title.str.contains('algo trading',case=False)]
data_contain_single_string.head(5)
```

```python
# find the rows which have 'python' or 'tactics' keywords in course_title column
search_values = ['tactics','python']
data_contain_multiple_keyword = data[data.course_title.str.contains('|'.join(search_values ),case=False)]
data_contain_multiple_keyword
```

```python
# find the rows which have 'TRADING TACTICS' or 'python web programming' or 'Python algo trading' strings in course_title column
search_values = ['TRADING TACTICS','python web programming','Python algo trading']
data_contain_multiple_string = data[data.course_title.str.contains('|'.join(search_values ),case=False)]
data_contain_multiple_string
```

## Find the similarity between two words

```python
# import spacy for finding word similarity using NLP
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
word1 = nlp('trading tactics')
word2 = nlp('finance practice')
```

```python
word1.similarity(word2)
```
